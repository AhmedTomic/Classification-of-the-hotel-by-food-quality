```python
data.set_index(["Positive_Review","Hotel_Name","lat","lng"],inplace=True)
```

```python
#Find email with word "word"
find_word=data.filter(like='food',axis=0)  
```

```python
find_word.head()
```

```python
selection=find_word.groupby(['Hotel_Name','Positive_Review','lat','lng']).size().to_frame()
```

```python
selection.head(33)
```

```python
selection.head()
```

```python
select=selection.sort_values(by=0,ascending=False).sort_values(by="Hotel_Name")
```

```python
select.head(10)
```

```python
filtering=select.groupby(['Hotel_Name','lat','lng']).sum()
```

```python
filtering.head()
```

```python
rank=filtering.sort_values(by=0,ascending=False)
```

```python
rank_10=rank.head(10)
```

```python
rank_10
```

```python
sf=rank_10
```

```python
sf
```

```python
type(sf)
```

```python
sf.reset_index(level=sf.index.names, inplace=True)
```

```python
sf.columns
```

```python
sf["lat"]
```

```python
sf["lng"]
```

```python
Hotel=sf["Hotel_Name"].values.tolist()
```

```python
Hotel
```

```python
locationlist = sf[["lat","lng"]].values.tolist()
```

```python
locationlist
```

```python
import folium
import os

#create a map
this_map = folium.Map(zoom_start=12,tiles='OpenStreetMap')

for point in range(len(locationlist)):
    folium.Marker((locationlist[point]), popup=Hotel[point]).add_to(this_map)


#Set the zoom to the maximum possible
this_map.fit_bounds(this_map.get_bounds())

#Save the map to an HTML file
#this_map.save(os.path.join('html_map_output/basic_pop_up.html'))

this_map.save(os.path.join('Hotel'))

this_map
```
